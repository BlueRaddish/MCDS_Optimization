```python
# Brute force
def return_dominating_set(G):
    """
    Find a dominating set using brute force.
    """
    for k in range(1, G.number_of_nodes() + 1):
        for nodes in itertools.combinations(G.nodes(), k):
            if is_connected_dominating_set(G, nodes):
                return nodes
    return None

def draw_dominating_set(G):
    """
    Draw the graph with the dominating set highlighted.
    """
    nodes = return_dominating_set(G)
    print(nodes)
    pos = {(x,y):(y,-x) for x,y in G.nodes()}
    plt.figure(figsize=(6,6))
    nx.draw(G, pos=pos, 
            node_color='lightgreen', 
            with_labels=True,
            node_size=600)
    
    # Highlight the dominating set
    nx.draw_networkx_nodes(G, pos, nodelist=nodes, node_color='orange')
    
    plt.show()
```
